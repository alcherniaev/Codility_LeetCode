## https://tastedive.com/read/api

```python
def get_movies_from_tastedive(name):
    params = {"q": name, "type": "movies", "limit": 5}
    response = requests.get("https://tastedive.com/api/similar", params=params)
    return json.loads(response.text)
```

```python
get_movies_from_tastedive("Bridesmaids")
```

## Maximum/Minimum Integer

```python
"""float('inf'), float('-inf') as the initial value of min_val and max_val"""
min_val, max_val = float('inf'), float('-inf')

for i in range(10):
    min_val = min(min_val, i)
    max_val = max(max_val, i)
print(f"min_val = {min_val}, \nmax_val = {max_val} ")
```
